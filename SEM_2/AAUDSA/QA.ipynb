```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline("text-generation", model="gpt2")
```

```python
context = """
Artificial Intelligence (AI) is the simulation of human intelligence in machines.
It enables machines to learn from data, adapt to new inputs, and perform tasks like humans.
AI is widely used in healthcare, finance, and robotics.
"""
```

```python
def ask_question(question):
    result = qa_pipeline(question=question, context=context)
    return result['answer']
```

```python
question = "Where is AI used?"

# Combine question + context
input_text = f"Context: {context}\nQuestion: {question}\nAnswer:"

result = qa_pipeline(input_text, max_new_tokens=30)
print(result[0]['generated_text'].split("Answer:")[-1])

print(result[0]['generated_text'])
```
