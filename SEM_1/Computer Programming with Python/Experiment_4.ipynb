### Operator Precedence, Conditional Statements and Loops

#### Objective
- To understand operator precedence in Python and to apply conditional statements and looping constructs(for and while)
-  for solving arithmetic expressions, dictionary operations, number reversal, and generating multiplication tables.

```python
"""
a.	Write a program to solve the expression: x = 10 + 2 * 3 ** 2 - 8 / 4 and show each expression calculation and output.  
"""
print("3 ** 2 =", 3 ** 2)
print("2 * 3 ** 2 =", 2 * 3 ** 2)
print("8 / 4 =", 8 / 4)
x = 10 + 2 * 3 ** 2 - 8 / 4
print("Final value of x =", x)
```

- The expression is evaluated step-by-step based on operator precedence.
- Exponentiation is performed first, followed by multiplication, division, addition, and subtraction.
- Each intermediate result is printed before calculating the final value.

```python
"""
b.	Write a Python program that creates a dictionary to store three studentsâ€™ names as keys and their marks as values, then:   
    i.	Prints the dictionary   
    ii.	Prints the marks of a student whose name is entered by the user  

"""

students = {
    "Aman": 85,
    "Riya": 90,
    "Karan": 78
}
print("Student Dictionary:", students)
name = input("Enter student name: ")

if name in students:
    print("Marks:", students[name])
else:
    print("Student not found")
```

- A dictionary stores student names as keys and marks as values.
- The complete dictionary is displayed.
- The user enters a student name, and the corresponding marks are retrieved.

```python
"""
c. Print the reverse of a number
"""
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed number:", rev)
```

- The given number is reversed using a while loop.
- Digits are extracted using modulo and integer division.
- The reversed number is constructed digit by digit.

```python
"""
d. Write the program to print multiplication table of n, where n is entered by the user.     
"""
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(n, "x", i, "=", n * i)
```

- A number is taken as input from the user.
- A for loop generates the multiplication table from 1 to 10.
