```python
"""
a.) Display data of different types using variables and literal constants
"""
a = 10
b = 3.14
c = "Hello Python"
d = True
print("Integer value:", a)
print("Float value:", b)
print("String value:", c)
print("Boolean value:", d)
print("Integer literal:", 25)
print("Float literal:", 7.5)
print("String literal:", "Welcome")
print("Boolean literal:", False)
```

```python
"""
b.) Application of += operator on strings
"""
str1 = "Hello"
str2 = " World"
str1 += str2
print("Resulting string:", str1)
```

- The += operator is used with strings.
- This operator appends the second string to the first string.
- It shows the use of compound assignment operators in Python.

```python
"""
c.) Addition and multiplication on string variables
"""
```

```python
s1 = "Python "
s2 = "Programming"

result_add = s1 + s2
result_mul = s1 * 3

print("Addition of strings:", result_add)
print("Multiplication of string:", result_mul)
```

- The + operator concatenates two strings.
- The * operator repeats a string multiple times.
- The program demonstrates string operations using arithmetic operators.
