### Operators based programs

#### Objective:
- To understand and apply arithmetic and assignment operators in Python by implementing programs for calculating the area of a triangle using 
Heron’s Formula and for finding the total amount of money in a piggy bank using different coin denominations.

```python
"""
a.) Area of a Triangle using Heron’s Formula
"""
a = float(input("Enter first side: "))
b = float(input("Enter second side: "))
c = float(input("Enter third side: "))
s = (a + b + c) / 2
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
print("Area of triangle:", area)
```

- The three sides of the triangle are taken as input from the user.
- The semi-perimeter is calculated using arithmetic operators.
- Heron’s Formula is applied to compute the area of the triangle.
- The exponentiation operator is used to calculate the square root.

```python
"""
b.) Total Amount of Money in a Piggy Bank
"""

rs10 = int(input("Enter number of Rs 10 coins: "))
rs5 = int(input("Enter number of Rs 5 coins: "))
rs2 = int(input("Enter number of Rs 2 coins: "))
rs1 = int(input("Enter number of Rs 1 coins: "))
total = (rs10 * 10) + (rs5 * 5) + (rs2 * 2) + (rs1 * 1)
print("Total amount in piggy bank: Rs", total)
```

- The number of coins of each denomination is accepted as input.
- Each coin count is multiplied by its respective value.
- The total amount is calculated by adding all the values.
