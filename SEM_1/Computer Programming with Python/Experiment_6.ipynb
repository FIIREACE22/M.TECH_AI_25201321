### Programs Based on Functions

#### Objective
- To understand the use of functions in Python, including lambda functions, recursive functions, and list processing using loops.

```python
"""
a) Write a Python program that uses a lambda function to: 
    i.	Take a number as input. 			
    ii.	Use a lambda function to square the number.
    iii.   Print the squared value. 
"""
n = int(input("Enter a number: "))
square = lambda x: x * x
print("Squared value:", square(n))
```

- A number is taken as input from the user.
- A lambda function is used to calculate the square of the number.
- The squared value is printed as output.

```python
"""
b.	Write a recursive function count digits(n) that returns the number of digits in a positive integer n. 
"""
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)

num = int(input("Enter a positive integer: "))
print("Number of digits:", count_digits(num))
```

- A recursive function is defined to count the digits of a positive integer.
- The function divides the number by 10 in each recursive call.
- The recursion ends when the number becomes zero.

```python
"""
c.	Write a Python program to: 
     i.	Take n integers into a list. 
    ii.	Print the list in original order. 
   iii. Print the list in reverse order (without using [::-1] or reverse()). 
"""
n = int(input("Enter number of elements: "))
lst = []
for i in range(n):
    lst.append(int(input("Enter element: ")))
print("Original list:", lst)
print("Reversed list:", end=" ")
for i in range(len(lst) - 1, -1, -1):
    print(lst[i], end=" ")
```

- A list of integers is created based on user input.
- The list is printed in its original order.
- The list is printed in reverse order using a loop without using built-in reverse methods.
