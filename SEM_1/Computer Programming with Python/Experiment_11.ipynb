### Programming Exercises with Classes, Objects and Inheritance
#### Objective
- To understand object-oriented programming concepts in Python such as classes, objects, class variables, and relationships between multiple classes through employee management and student–course–department association

```python
"""
a.	Write a program with class Employee that keeps a track of the number of employees in an organization 
and also stores their name, designation and salary details.  
"""
class Employee:
    count = 0

    def __init__(self, name, designation, salary):
        self.name = name
        self.designation = designation
        self.salary = salary
        Employee.count += 1

    def display(self):
        print("Name:", self.name)
        print("Designation:", self.designation)
        print("Salary:", self.salary)


e1 = Employee("Aman", "Manager", 50000)
e2 = Employee("Riya", "Developer", 40000)

e1.display()
e2.display()

print("Total Employees:", Employee.count)
```

- class Employee is defined with a class variable to track the number of employees.
- Each time an object is created, the employee count is incremented.
- Employee details such as name, designation, and salary are stored and displayed.
