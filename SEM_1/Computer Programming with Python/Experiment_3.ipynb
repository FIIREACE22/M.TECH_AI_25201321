### String Manipulation, Number System and Conversions

#### Objective
- To understand string manipulation and number system conversions in Python by converting a decimal number into binary, 
octal, and hexadecimal formats and by performing arithmetic operations after converting a string into a floating-point number.

```python
"""
a.) Decimal to Binary, Octal and Hexadecimal Conversion
"""
num = int(input("Enter a decimal number: "))
print("Binary:", bin(num))
print("Octal:", oct(num))
print("Hexadecimal:", hex(num))
```

- An integer value in decimal format is taken as input from the user.
- Built-in functions bin(), oct(), and hex() are used to convert the number into binary, octal, and hexadecimal formats respectively.
- The converted values are displayed as output.

```python
"""
b.) String to Float Conversion and Arithmetic Operation
"""
num_str = input("Enter a floating point number: ")
num = float(num_str)
result = num + 10.5
print("Result:", result)
```

- A string representing a floating-point number is read from the user.
- The string is converted into a float using the float() function.
- An arithmetic operation is performed by adding 10.5 to the converted value.
- The final result is printed.
