```python
def find_smallest_multiple(n: int) -> int:
    
    if n == 1:
        return 1

    x = 1  

    while True:
        divisible = True

        for i in range(1, n + 1):
            if x % i != 0:
                divisible = False
                break

        if divisible:
            return x

        x += 1  

print(find_smallest_multiple(10))
```

```python
def cubesum(n: int) -> int:
   
    total = 0
    while n > 0:
        digit = n % 10
        total += digit ** 3
        n //= 10
    return total


def isArmstrong(n: int) -> bool:
    
    return cubesum(n) == n


def PrintArmstrong(limit: int):

    print(f"Armstrong numbers up to {limit}:")
    for num in range(1, limit + 1):
        if isArmstrong(num):
            print(num)

PrintArmstrong(1000)

print(isArmstrong(153))   
print(isArmstrong(370))   
print(isArmstrong(200))   
```

```python
def to_binary(n):
    if n == 0:
        return "0"
    result = ""
    while n > 0:
        result = str(n % 2) + result
        n //= 2
    return result


def to_octal(n):
    if n == 0:
        return "0"
    result = ""
    while n > 0:
        result = str(n % 8) + result
        n //= 8
    return result


def to_hexadecimal(n):
    if n == 0:
        return "0"
    hex_digits = "0123456789ABCDEF"
    result = ""
    while n > 0:
        result = hex_digits[n % 16] + result
        n //= 16
    return result


num = int(input("Enter a decimal number: "))

print("Binary equivalent      :", to_binary(num))
print("Octal equivalent       :", to_octal(num))
print("Hexadecimal equivalent :", to_hexadecimal(num))
```

```python
def print_square_list():
    squares = []  
    for i in range(1, 31):  
        squares.append(i * i)

    print("List of squares from 1 to 30:")
    print(squares)


print_square_list()
```

```python
sentence = input("Enter a sentence: ")

vowels = 0
consonants = 0
digits = 0
special = 0

for ch in sentence:
    if ch.isalpha():  
        if ch.lower() in "aeiou":
            vowels += 1
        else:
            consonants += 1
    elif ch.isdigit():  
        digits += 1
    elif ch != " ":  
        special += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special characters:", special)
```

```python

students = {}

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    students[name] = marks


highest_student = max(students, key=students.get)
lowest_student = min(students, key=students.get)

print("\nStudent with highest marks:")
print(highest_student, ":", students[highest_student])

print("\nStudent with lowest marks:")
print(lowest_student, ":", students[lowest_student])
```
