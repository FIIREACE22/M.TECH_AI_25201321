#### Objective
- To understand basic data structures in Python such as lists, tuples, and sets by performing operations like removing duplicates, searching elements, and finding common and unique elements between sequences.

```python
"""
Write a Python program to: 
    i.	Take n integers from the user and store them in a list. 
   ii.	Create a new list that contains the same elements but without duplicates. 
  iii.	Print both the original list and the new list.  
"""
n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    lst.append(int(input("Enter element: ")))

unique_list = list(set(lst))

print("Original list:", lst)
print("List without duplicates:", unique_list)
```

- A list of integers is created using user input.
- A new list is formed by converting the list into a set to remove duplicates.
- Both the original and duplicate-free lists are printed.

```python
"""
Write a Python program to: 
    i.	Create a tuple, e.g. nums = (10, 20, 30, 40, 50). 
   ii.	Ask the user to enter a number. 
  iii.	Print the index of that number in the tuple if it exists, otherwise print a message saying it is not present. 
"""
nums = (10, 20, 30, 40, 50)

x = int(input("Enter a number: "))

if x in nums:
    print("Index:", nums.index(x))
else:
    print("Number not present")
```

- A tuple with predefined elements is created.
- The user enters a number to search.
- If the number exists, its index is printed; otherwise, a message is displayed.

```python
"""
Write a program that reads two sequences of integers (allow duplicates) and prints:
    i.	The set of elements common to both sequences, 
   ii.	The set of elements that appear in exactly one of the two sequences (symmetric difference). Elements in each 
        output set should be printed in sorted order without duplicates.
"""
n1 = int(input("Enter number of elements in first sequence: "))
seq1 = []

for i in range(n1):
    seq1.append(int(input("Enter element: ")))

n2 = int(input("Enter number of elements in second sequence: "))
seq2 = []

for i in range(n2):
    seq2.append(int(input("Enter element: ")))

set1 = set(seq1)
set2 = set(seq2)

print("Common elements:", sorted(set1 & set2))
print("Symmetric difference:", sorted(set1 ^ set2))
```

- Two integer sequences are read from the user.
- Sets are created to remove duplicates automatically.
- Common elements and symmetric difference are computed using set operations.
- The results are printed in sorted order.
