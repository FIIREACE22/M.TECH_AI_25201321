### File Handling
#### Objective
- To understand file handling operations in Python by counting specific characters in a text file and copying a fixed number of bytes from one binary file to another.

```python
"""
a.	Write a program that counts the number of tabs, spaces and newline characters in a file. 
"""
file = open("input.txt", "r")
text = file.read()
tabs = text.count("\t")
spaces = text.count(" ")
newlines = text.count("\n")
file.close()
print("Tabs:", tabs)
print("Spaces:", spaces)
print("Newlines:", newlines)
```

- The file is opened in read mode.
- The entire content of the file is read as a string.
- Tabs, spaces, and newline characters are counted using string functions.
- The results are printed after closing the file.

```python
"""
b.	Write a program that copies first 10 bytes of a binary file into another.
""" 
source = open("source.bin", "rb")
destination = open("destination.bin", "wb")

data = source.read(10)
destination.write(data)

source.close()
destination.close()

print("First 10 bytes copied successfully")
```

- The source file is opened in binary read mode.
- The destination file is opened in binary write mode.
- The first 10 bytes are read from the source file and written into the destination file.
- Both files are closed after the operation.
