### String and Its Operations

#### Objective
- To understand string manipulation in Python by performing vowel removal, palindrome checking, and word order reversal using string operations and control structures.

```python
"""
a.	Write a program that takes a string as input and prints a new string where all vowels are removed, 
but the order of the remaining characters is preserved.
"""
s = input("Enter a string: ")
result = ""

for ch in s:
    if ch.lower() not in "aeiou":
        result += ch

print("String without vowels:", result)
```

- The input string is traversed character by character.
- Vowels are identified and removed while preserving the original order of remaining characters.
- The resulting string is printed.

```python
"""
b.	Given a string, write a program to check whether it is a palindrome or not, 
ignoring case and all non-alphanumeric characters.
"""
s = input("Enter a string: ")
cleaned = ""

for ch in s:
    if ch.isalnum():
        cleaned += ch.lower()

if cleaned == cleaned[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

- The string is cleaned by removing non-alphanumeric characters and converting all characters to lowercase.
- The cleaned string is compared with its reverse.
- Based on the comparison, the program determines whether the string is a palindrome.

```python
"""
c.	Write a program that takes a sentence as input and prints the words in reverse order, while keeping 
the characters inside each word in the same order.
"""
sentence = input("Enter a sentence: ")
words = sentence.split()

for i in range(len(words) - 1, -1, -1):
    print(words[i], end=" ")
```

- The string is cleaned by removing non-alphanumeric characters and converting all characters to lowercase.
- The cleaned string is compared with its reverse.
- Based on the comparison, the program determines whether the string is a palindrome.
