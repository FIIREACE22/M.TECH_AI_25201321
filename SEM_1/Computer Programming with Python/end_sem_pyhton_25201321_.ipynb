1.Write a program to calculate salary of an employee given his basic pay, HRA =10per
cent of basic pay, TA=5 per cent of basic pay. Define HRA and Ta as constants and use them
to calculate the salary of the employee.

```python
# Inputs
basic_pay = float(input("Enter Basic Pay==> "))

# Calculations
hra_amount =  0.10 * basic_pay
ta_amount =  0.05 * basic_pay
salary = basic_pay + hra_amount + ta_amount

# Output
print("HRA==>", hra_amount)
print("TA==>", ta_amount)
print("Total Salary==>", salary)
```

2.A university maintains student fee records across multiple departments. Due to frequentpolicy changes, incorrect updates have occurred, leading to audit failures.
Design a Python system that performs the following:
a. Define a Student class storing:
o Roll Number (int)
o Name (string)

o Department (string)
o Pending Fee (int)
o CGPA (float)
b. Read student records from a text file students.txt.
c.. Separate students into two output files:
o even_roll.txt → students with even roll numbers
o odd_roll.txt → students with odd roll numbers
d. Reopen both files and update records such that:
o If Pending Fee &lt; 5000 AND CGPA &gt; 8.0, set Pending Fee to 0
o This update must not use class objects or class methods
e. Maintain an audit log file that records:
o Original fee
o Updated fee
o Roll number
o Timestamp
f. Implement:
o Custom exceptions for invalid data
o Nested try–except
o Re-throw exceptions after logging
o Graceful program termination

```python
import datetime

# Custom exception for wrong data
class InvalidStudentDataError(Exception):
    pass


# Student class
class Student:
    def __init__(self, roll_no, name, department, pending_fee, cgpa):
        self.roll_no = roll_no
        self.name = name
        self.department = department
        self.pending_fee = pending_fee
        self.cgpa = cgpa


# Function to read student details
def load_students(file_name):
    student_list = []

    try:
        f = open(file_name, "r")
        for line in f:
            try:
                data = line.strip().split(",")

                if len(data) != 5:
                    raise InvalidStudentDataError("Invalid record format")

                roll_no = int(data[0])
                name = data[1]
                department = data[2]
                pending_fee = int(data[3])
                cgpa = float(data[4])

                student_list.append(
                    Student(roll_no, name, department, pending_fee, cgpa)
                )

            except Exception as err:
                write_error_log(err)
                raise   # rethrow after logging

        f.close()

    except FileNotFoundError:
        print("Student file not found.")
        exit()

    return student_list


# Split students based on roll number
def write_even_odd_files(students):
    even_file = open("even_roll.txt", "w")
    odd_file = open("odd_roll.txt", "w")

    for s in students:
        line = f"{s.roll_no},{s.name},{s.department},{s.pending_fee},{s.cgpa}\n"
        if s.roll_no % 2 == 0:
            even_file.write(line)
        else:
            odd_file.write(line)

    even_file.close()
    odd_file.close()


# Log audit details
def write_audit_log(roll, old_fee, new_fee):
    log = open("audit_log.txt", "a")
    time_now = datetime.datetime.now()
    log.write(f"{roll},{old_fee},{new_fee},{time_now}\n")
    log.close()


# Log errors
def write_error_log(error):
    log = open("error_log.txt", "a")
    log.write(f"{datetime.datetime.now()} - {error}\n")
    log.close()


# Update fees without using Student objects
def update_fee_file(file_name):
    new_data = []

    try:
        f = open(file_name, "r")
        for line in f:
            try:
                data = line.strip().split(",")

                roll = int(data[0])
                fee = int(data[3])
                cgpa = float(data[4])

                old_fee = fee
                if fee < 5000 and cgpa > 8.0:
                    fee = 0
                    write_audit_log(roll, old_fee, fee)

                data[3] = str(fee)
                new_data.append(",".join(data))

            except Exception as err:
                write_error_log(err)
                raise

        f.close()

        f = open(file_name, "w")
        for record in new_data:
            f.write(record + "\n")
        f.close()

    except Exception:
        print(f"Problem while updating {file_name}")
        exit()


# Main program
def main():
    try:
        students = load_students("students.txt")
        write_even_odd_files(students)

        update_fee_file("even_roll.txt")
        update_fee_file("odd_roll.txt")

        print("Student records processed successfully.")


    except Exception:
        print("Program stopped due to an unexpected error.")


main()
```

3.Write a Python program that:
 Accepts an integer n
 Prints a hollow butterfly pattern
 Wings must be symmetric
 Center must be hollow
 Borders must be filled

```python
n = int(input("Enter number of rows: "))

# Top part of butterfly
for row in range(1, n + 1):

    # Left wing
    for col in range(1, row + 1):
        if col == 1 or col == row:
            print("*", end="")
        else:
            print(" ", end="")

    # Space between wings
    gap = 2 * (n - row)
    print(" " * gap, end="")

    # Right wing
    for col in range(1, row + 1):
        if col == 1 or col == row:
            print("*", end="")
        else:
            print(" ", end="")
    print()


# Bottom part of butterfly
for row in range(n, 0, -1):

    # Left wing
    for col in range(1, row + 1):
        if col == 1 or col == row:
            print("*", end="")
        else:
            print(" ", end="")

    # Space between wings
    gap = 2 * (n - row)
    print(" " * gap, end="")

    # Right wing
    for col in range(1, row + 1):
        if col == 1 or col == row:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
