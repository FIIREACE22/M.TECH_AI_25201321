### Operator Overloading and Error Handling
#### Objective
- To understand operator overloading using special methods in Python and to apply error handling techniques for input validation and exception management.

```python
"""
a.	Write a program that overloads the + operator on a class Student that has attributes name and marks.   
"""
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks


s1 = Student("Aman", 80)
s2 = Student("Riya", 85)

print("Total Marks:", s1 + s2)
```

- A class Student is defined with attributes name and marks.
- The + operator is overloaded using the __add__() method.
- Adding two Student objects returns the sum of their marks.

```python
"""
b.	Write a program that validates name and age as entered by the user 
    to determine whether the person can cast vote or not.  
"""
name = input("Enter name: ")
age = int(input("Enter age: "))

if age >= 18:
    print(name, "is eligible to vote")
else:
    print(name, "is not eligible to vote")
```

- Name and age are taken as input from the user.
- A conditional statement checks voting eligibility based on age.
- The result is displayed accordingly.

```python
"""
c. Write a program that asks the user to enter an integer. If the user enters something that is not an integer (like 
"abc" or 3.5), the program should catch the error, print "Invalid input, please enter an integer.", and ask again 
until a valid integer is entered.
"""
while True:
    try:
        n = int(input("Enter an integer: "))
        print("Valid integer entered:", n)
        break
    except ValueError:
        print("Invalid input, please enter an integer.")
```

- The program repeatedly asks the user for input.
- If the input is not an integer, a ValueError is caught.
- The user is prompted again until a valid integer is entered.
