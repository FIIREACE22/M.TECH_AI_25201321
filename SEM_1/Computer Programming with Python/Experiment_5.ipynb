### Nested Loops, String Multiplication and Keyword Arguments

#### Objective
- To understand the use of nested loops, string multiplication, keyword arguments in functions, and built-in mathematical operations in Python through pattern printing, multiplication tables, function calls, and numeric computations.

```python
"""
a.	Write a program to print the following pattern:   
       1   
  	  12   
  	 123   
    1234  
   12345 
"""
n = 5

for i in range(1, n + 1):
    for space in range(n - i):
        print(" ", end="")
    for num in range(1, i + 1):
        print(num, end="")
    print()
```

- The program uses nested loops to print a right-aligned number pattern.
- The outer loop controls the total number of rows.
- The first inner loop prints spaces to align the numbers to the right.
- The second inner loop prints numbers from 1 up to the current row number.
- After each row is printed, the program moves to the next line to complete the pattern.

```python
"""
b.	Write the program to print multiplication table of n, where n is entered by the user 
"""
n = int(input("Enter a number: "))
for i in range(1, 11):
    print(n, "x", i, "=", n * i)
```

- A number is taken as input from the user.
- A for loop generates the multiplication table from 1 to 10.

```python
"""
c.	Write a Python function student info that takes three keyword arguments: name, age, and course. 
Call the function using keyword arguments in any order and print the information in a readable format. 
"""
def student_info(name, age, course):
    print("Name:", name)
    print("Age:", age)
    print("Course:", course)

student_info(course="AI", name="Aman", age=21)
```

- A function is defined with keyword arguments.
- The function is called by passing arguments in a different order.
- The output is printed in a readable format.

```python
"""
d.	Write a program that prints absolute value, square root and cube of a 
number. 
"""
num = float(input("Enter a number: "))

print("Absolute value:", abs(num))
print("Square root:", num ** 0.5)
print("Cube:", num ** 3)
```

- Built-in and arithmetic operators are used to calculate absolute value, square root, and cube of a number.
- The results are printed directly.
