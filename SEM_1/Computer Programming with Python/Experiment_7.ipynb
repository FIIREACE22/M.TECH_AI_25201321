### Modules and Packages

#### Objective
- To understand the use of Python modules and packages by importing built-in and external modules using aliases and applying them for random number generation and data visualization.

```python
"""
a.	Write a Python program that: 
    i.	Imports the random module with alias name rnd. 
    ii.	Prints 5 random integers between 1 and 100 using the alias. 
"""
import random as rnd
for i in range(5):
    print(rnd.randint(1, 100))
```

- The random module is imported using the alias rnd.
- The randint() function is used to generate random integers.
- Five random numbers between 1 and 100 are printed.

```python
"""
b.	Write a Python program that: 
    i. Imports pyplot from the matplotlib package as plt.
    ii.     Plots a simple line graph for x = [1, 2, 3, 4] and y = [1, 4, 9, 16] 
"""
from matplotlib import pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.figure(figsize=(5, 3))
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Graph")
plt.show()
```

- The pyplot module is imported from the matplotlib package using the alias plt.
- Two lists representing x and y values are defined.
- A simple line graph is plotted using the plot() function.
- Labels and title are added and the graph is displayed.
